```python
#FEW SHOT
!pip install transformers

from transformers import pipeline
translation =pipeline("translation_en_to_es",model="google/flan-t5-base")
prompt="""Translate to spanish:
English: Good morning
Spanish: buen día
English: How are you?
Spanish:
give me only answer
"""
print(translation(prompt))
```
